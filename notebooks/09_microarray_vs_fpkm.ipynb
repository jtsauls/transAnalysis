## Import data

The goal is to get a dict with gene names, in bxxxx format, as keys, and unnormalized microarray levels as values.

```python
# Import file
filename = 'data/palsson_microarray_data_w_info_080826.xls'
sheetname = 'GenExpVal'
ma_df_root = pd.read_excel(filename, sheetname, index_col=None)
```

```python
ma_df = ma_df_root
# From the excel file, make a dictionary of genes to data. 
exp_name = 'ec_aer_wild_O_a_SciA.CEL'  # Name of column in excel, glucose aerobic
#exp_name = 'ec_aer_wild_nO_a_SciA.CEL'  # glucose anaerobic
# Pull out genes and values for experiment of concern
ma_df = ma_df.loc[:, ['genes', exp_name]]  
# Only use rows with gene names containing 'bxxxx'
ma_df = ma_df[ma_df.genes.str.contains('b\d\d\d\d')]
HTML(ma_df.to_html())
```

```python
# Reg ex so names are in form bxxxx
gene_names = list(ma_df.genes)
regexp = re.compile('b\d\d\d\d')
genes = [m.group(0) for l in gene_names for m in [regexp.search(l)] if m]

# Remove unicode
genes = [gene.encode('ascii', 'ignore') for gene in genes]

# Data log2 normalized, reverse this
ma_raw = np.power(ma_df.loc[:, exp_name], 2)
#ma_raw = ma_df.loc[:, exp_name]

# Check for duplicates
if len(gene_names) != len(set(genes)):
    print('Warning, non-unique genes in list')
else:
    gene_ma_dict = dict(zip(genes, ma_raw))
```

## Compare microarray data with RNAseq data

How correlated are they?

```python
fpkm_filename = 'data/bop27_1_isoforms.fpkm_tracking'
#fpkm_filename = filenameroot + 'bop27_2_isoforms.fpkm_tracking' 
#fpkm_filename = 'data/anaerobe1_isoforms.fpkm_tracking'
#fpkm_filename = filenameroot + 'anaerobe2_isoforms.fpkm_tracking'
#fpkm_filename = 'data/rpoBisoforms.fpkm_tracking'

# Creat the dictionary linking fpkms to reactions. 
gene_fpkm_dict = ta.make_gene_fpkm_dict(fpkm_filename)
reaction_fpkm_dict = ta.make_reaction_fpkm_dict(model, gene_fpkm_dict)
```

### Compare reaction transcript levels in the two datasets

This takes reactions with no transcript information to have a level of zero. This comparison, because each dataset is mapped to the same model, avoids the problem that each method, RNAseq and microarray, do not cover the same genes. 

```python
# Pull out reaction transcript levels from dicts to plot
reaction_levels = [0, 0]

# Change (value, reaction) tuple structure to two lists and
# pull out values.
reaction_fpkm_list = reaction_fpkm_dict.items()
reaction_levels[0] = zip(*reaction_fpkm_list)[1]
reaction_levels[0] = np.array([x if ~np.isnan(x) else 0 for 
                      x in reaction_levels[0]])
#reaction_levels[0] = reaction_levels[0] / max(reaction_levels[0])
a = min(reaction_levels[0])
b = max(reaction_levels[0])
reaction_levels[0] = np.log2(1 + 1023 * ((reaction_levels[0] - a) / (b - a)))
#reaction_levels[0] = np.log2(reaction_levels[0])
reaction_levels[0] = np.array([x if x > 0 else 0 for 
                      x in reaction_levels[0]])

reaction_ma_list = reaction_ma_dict.items()
reaction_levels[1] = zip(*reaction_ma_list)[1]
reaction_levels[1] = np.array([x if ~np.isnan(x) else 0 for
                      x in reaction_levels[1]])
#reaction_levels[1] = reaction_levels[1] / max(reaction_levels[1])
a = min(reaction_levels[1])
b = max(reaction_levels[1])
reaction_levels[1] = np.log2(1 + 1023 * ((reaction_levels[1] - a) / (b - a)))
#reaction_levels[1] = np.log2(reaction_levels[1])
reaction_levels[1] = np.array([x if x > 0 else 0 for 
                      x in reaction_levels[1]])
```

```python
# Correlation
pear_cor = stats.pearsonr(reaction_levels[0],
                          reaction_levels[1])
print("Pearson correlation, p-value", pear_cor)

spear_cor = stats.spearmanr(reaction_levels[0],
                            reaction_levels[1])
print("Spearman correlation, p-value", spear_cor)

# Plotting
plt.figure(figsize=(10,10), dpi=100)
# Value comparison.
plt.title('RNAseq vs. microarray levels for matched reactions')
plt.xlabel('FPKM (log2 normalized)')
plt.ylabel('mircroarray levels (log2 normalized)') 
plt.scatter(reaction_levels[0], reaction_levels[1], c='g')
```

### Consensus between genes measures via microarray and RNAseq

```python
print('There are', len(gene_fpkm_dict), 'genes in the seq data.')
print('There are', len(gene_ma_dict), 'genes in the ma data.')
```

```python
# What genes are in one and not the other
inseq_notma = []
for gene_name in gene_fpkm_dict.keys():
    if gene_name not in gene_ma_dict.keys():
        inseq_notma.append(gene_name)
        
inma_notseq = []
for gene_name in gene_ma_dict.keys():
    if gene_name not in gene_fpkm_dict.keys():
        inma_notseq.append(gene_name)
```

```python
print(len(inseq_notma))
print(len(inma_notseq))
```

```python
# Find consesus set and make consensus dicts
consensus = set(gene_ma_dict.keys()).intersection(gene_fpkm_dict.keys())
print(len(consensus))
ma_con_dict = dict()
seq_con_dict = dict()
for gene in consensus:
    ma_con_dict[gene] = gene_ma_dict[gene]
    seq_con_dict[gene] = gene_fpkm_dict[gene]
```

```python
# But does it matter? Are all the genes from iJO accounted for?
no_data_genes_ma = []
no_data_genes_fpkm = []
for gene in model.genes:
    gene_name = gene.id
    try:
        gene_ma_dict[gene_name]
    except:
        no_data_genes_ma.append(gene_name)
    try:
        gene_fpkm_dict[gene_name]
    except:
        no_data_genes_fpkm.append(gene_name)
```
