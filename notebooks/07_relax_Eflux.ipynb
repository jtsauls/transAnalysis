```python
Eflux_bottlenecks = []

for reaction in Eflux_model.reactions:
    reaction_name = reaction.id
    if Eflux_model.solution.x_dict[reaction_name] != 0.0 and \
    (Eflux_model.solution.x_dict[reaction_name] == \
    Eflux_model.reactions.get_by_id(reaction_name).upper_bound or \
    Eflux_model.solution.x_dict[reaction_name] == \
    Eflux_model.reactions.get_by_id(reaction_name).lower_bound):
        Eflux_bottlenecks.append((reaction_name,
            Eflux_model.solution.x_dict[reaction_name],
            Eflux_model.reactions.get_by_id(reaction_name).lower_bound,
            Eflux_model.reactions.get_by_id(reaction_name).upper_bound,
            model_root.reactions.get_by_id(reaction_name).lower_bound,
            model_root.reactions.get_by_id(reaction_name).upper_bound,
            reaction_fpkm_dict[reaction_name]))

print('Number of bottlenecks', len(Eflux_bottlenecks))
#print(Eflux_bottlenecks)
```

## Loosen bounds one by one

Below, we see how loosening bound for each reaction in the Eflux set up affects flux for that reaction and total flux. This only loosens the bottlenecks to minimize calculations

```python
Eflux_model = model_root.copy()
ta.Eflux(Eflux_model, reaction_fpkm_dict)

# Test the impact of Eflux bounds
f_diff = []
unbound_reactions = []
count = 0

relaxed_value = Eflux_model.reactions.Ec_biomass_iJO1366_core_53p95M.upper_bound

# Eflux orginal solution
Eflux_sol = Eflux_model.solution

for bottleneck in Eflux_bn:
    rxn = Eflux_model.reactions.get_by_id(bottleneck)
    
    # Save current bounds.
    ub_Eflux = Eflux_model.reactions.get_by_id(rxn.id).upper_bound
    lb_Eflux = Eflux_model.reactions.get_by_id(rxn.id).lower_bound
    
    # Change bounds back to that of original model
    Eflux_model.reactions.get_by_id(rxn.id).upper_bound = relaxed_value
    Eflux_model.reactions.get_by_id(rxn.id).lower_bound = -relaxed_value
        
    # Calculate new flux
    Eflux_model.optimize()
    
    # Save data for when f was different
    if abs(Eflux_model.solution.f - Eflux_sol.f) > 0.001:
        unbound_reactions.append((rxn.id,
                                  Eflux_model.solution.f - Eflux_sol.f))

    # Collecd data in massive list of tuples
    f_diff.append((rxn.id,
            Eflux_model.solution.f - Eflux_sol.f,
            Eflux_model.solution.x_dict[rxn.id] - Eflux_sol.x_dict[rxn.id],
            Eflux_model.solution.x_dict[rxn.id],
            Eflux_sol.x_dict[rxn.id],
            reaction_fpkm_dict[rxn.id],
            lb_Eflux,
            ub_Eflux))
    
    # Revert bounds to normal for next iteration
    Eflux_model.reactions.get_by_id(rxn.id).upper_bound = ub_Eflux
    Eflux_model.reactions.get_by_id(rxn.id).lower_bound = lb_Eflux
```

## Now do a comprehensive releasing of bounds

Loop through reactions, release the bounds and compute flux. For the reaction with the biggest difference in f, leave those bounds open and repeat the process until f approaches its non-Eflux growth-rate.

```python
# Combine and do iteratively, only saving the biggest difference. 
Eflux_model = model_root.copy()
ta.Eflux(Eflux_model, reaction_fpkm_dict)

# Eflux orginal solution
Eflux_sol = Eflux_model.solution

new_f = -1
old_f = Eflux_model.solution.f
previous_f = 0
relaxed_value = Eflux_model.reactions.Ec_biomass_iJO1366_core_53p95M.upper_bound

count = 0

# Stop when loosening produces f greater that normal FBA sol or solution doesn't change
while new_f != previous_f:
    count += 1;
    previous_f = new_f
    
    # Find bottlenecks for candidate reactions
    Eflux_bottlenecks = []
    for rxn in Eflux_model.reactions:
        rxn_name = reaction.id
        if Eflux_model.solution.x_dict[rxn_name] != 0.0 and \
        (Eflux_model.solution.x_dict[rxn_name] == rxn.upper_bound or \
        Eflux_model.solution.x_dict[rxn_name] == rxn.lower_bound):
            Eflux_bottlenecks.append(rxn_name)
            
    # Reduce bounds on bottlenecks
    for bottleneck in Eflux_bn:
        rxn = Eflux_model.reactions.get_by_id(bottleneck)
        
        # Save current bounds.
        ub_Eflux = Eflux_model.reactions.get_by_id(rxn.id).upper_bound
        lb_Eflux = Eflux_model.reactions.get_by_id(rxn.id).lower_bound
        
        # Loosen bounds
        Eflux_model.reactions.get_by_id(rxn.id).upper_bound = relaxed_value
        Eflux_model.reactions.get_by_id(rxn.id).lower_bound = -relaxed_value
            
        # Calculate new flux and find largest difference in f
        Eflux_model.optimize()
        temp_f = Eflux_model.solution.f
        
        # Save data for when f was different
        if temp_f > old_f:
            bn_temp = rxn.id
            old_f = temp_f
        
        # Revert bounds to normal for next iteration
        Eflux_model.reactions.get_by_id(rxn.id).upper_bound = ub_Eflux
        Eflux_model.reactions.get_by_id(rxn.id).lower_bound = lb_Eflux
        
    # Adjust model for bottleneck with biggest diff in f
    print('Round', count, 'Bottlneck', bn_temp, 'f', old_f)
    Eflux_model.reactions.get_by_id(bn_temp).upper_bound = relaxed_value
    Eflux_model.reactions.get_by_id(bn_temp).lower_bound = -relaxed_value
    Eflux_model.optimize()
    
    new_f = Eflux_model.solution.f  # Should be same as temp_f
    
    #if new_f > scaled_f:
    #   break
```
