### Two Dictionaries

Dictionary 1, minspan_k_dict, has minspan_ids as keys, which is currently just a number, and a list of reactions involved in that minspan as the values. There are 750 minspans. 

Dictionary 2, reaction_k_dict, has reaction ids as keys, and the id (number) of the minspans in which it is involved. There are 2583 reactions.

```python
# Dict 1, minspan id (aka number, as key, list of reactons as values
minspan_k_dict = ta.make_minspan_k_dict(minspan_list)

# Dict 2, reaction name as k, with minspans involved as values.
reaction_k_dict, reactions_wo_minspans = \
    ta.make_reaction_k_dict(model, minspan_k_dict, report_wo=True)
#reaction_k_dict = ta.make_reaction_k_dict1(minspan_list)
```

##Histograms of connectedness

```python
# How many minspans does each reaction participate in?
minspan_num_per_reaction_dict = dict()
for k, v in reaction_k_dict.iteritems():
    minspan_num_per_reaction_dict[k] = len(v)
    
max_connections = max(minspan_num_per_reaction_dict.values())

print('Most connected reaction w/', max_connections)
```

```python
# Plot
plt.figure(figsize=(10,10), dpi=100)
plt.hist(minspan_num_per_reaction_dict.values(), bins=max_connections)
#plt.xlim([1,50])
plt.ylim([0,50])
plt.title('Histogram of number of minspans per reaction')
plt.ylabel('Frequency (# Reactions in ... minspans)')
plt.xlabel('Reaction promiscuity, binsize=1')
plt.show()
```

```python
# How many reactions are in each minspan?
reaction_num_per_minspan_dict = dict()
for k, v in minspan_k_dict.iteritems():
    reaction_num_per_minspan_dict[k] = len(v)
    
max_participants = max(reaction_num_per_minspan_dict.values())
print('Biggest minspan', max_participants)
```

```python
# Plot
plt.figure(figsize=(10,10), dpi=100)
plt.hist(reaction_num_per_minspan_dict.values(), bins=max_participants)
plt.xlim([0,max_participants+5])
plt.title('Histogram of number of reactions per minspan')
plt.ylabel('Frequency (# Minspans with size ...)')
plt.xlabel('Minspan size, binsize=1')
plt.show()
```

## Which reactions are most connected, and what are thier FPKM values?

```python
reaction_proms = pd.Series(minspan_num_per_reaction_dict)
reaction_fpkms = pd.Series(reaction_fpkm_dict)

reaction_prom_df = {'In # minspans' : reaction_proms,
                    'FPKM' : reaction_fpkms}
columns = ['In # minspans', 'FPKM']
reaction_prom_df = pd.DataFrame(reaction_prom_df, 
                                columns=columns)
reaction_prom_df.sort('In # minspans', ascending=False, inplace=True)
HTML(reaction_prom_df.to_html(float_format=lambda x: '%10.0f' % x))
```

```python
# Graph connectetness vs FPKM value
connects = list(reaction_prom_df.loc[:,'In # minspans'])
fpkms = np.array(reaction_prom_df.loc[:,'FPKM'])
fpkms = np.log2(1 + ((fpkms - min(fpkms)) / (max(fpkms) - min(fpkms))))

plt.figure(figsize=(10,10), dpi=100)
plt.scatter(connects, fpkms)
plt.xlim([0, max(connects)+5])
plt.ylim([0, max(fpkms)])
plt.title('Promiscuity of reaction vs FPKM')
plt.xlabel('# Of minspans reaction is in')
plt.ylabel('Lognorm of FPKM of reaction')
plt.show()
```

## Calculating minspan ranks

The rank is the 'average' fpkm value for the minspan as a whole. Using normal aerobic data for now. 

```python
# Print dataframe of minspans. Index is ID
minspan_reactions = pd.Series(minspan_k_dict)
minspan_size = pd.Series(reaction_num_per_minspan_dict)
minspan_k_df = {'Reactions' : minspan_reactions,
                '# of Reactions' : minspan_size}
columns = ['Reactions', '# of Reactions']
minspan_k_df = pd.DataFrame(minspan_k_df, 
                            columns=columns)
```

```python
# Sort the minspans averages and add them to the df
minspan_k_df.sort(columns='Reactions', ascending=True, inplace=True)

# Rank the minspans straight up
minspan_fpkm_list = ta.make_minspan_fpkm_list(minspan_list,
                                              reaction_fpkm_dict,
                                              sort=True)

# You have to sort this list of tuples by name, as the df is sorted this way,
# to be able to cleanly add them. 
minspan_fpkm_list = sorted(minspan_fpkm_list,key=lambda x: x[1])
fpkm_per_minspan = zip(*minspan_fpkm_list)[0]

# We can only add this column like this because the 
minspan_k_df['FPKM avg'] = fpkm_per_minspan

# Sort and add rank
minspan_k_df.sort(columns='FPKM avg',
                  ascending=False, inplace=True)
```

```python
HTML(minspan_k_df.to_html())
```

```python
# Graph minspan size vs FPKM value
sizes = list(minspan_k_df.loc[:,'# of Reactions'])
fpkm_avgs = np.array(minspan_k_df.loc[:,'FPKM avg'])
fpkm_avgs = np.log2(1 + ((fpkm_avgs - min(fpkm_avgs)) \
                          / (max(fpkm_avgs) - min(fpkm_avgs))))

plt.figure(figsize=(10,10), dpi=100)
plt.scatter(sizes, fpkm_avgs)
plt.xlim([0, max(sizes)+5])
plt.ylim([0, max(fpkm_avgs)])
plt.title('Size of minspan vs FPKM avg')
plt.xlabel('# Of reactions in minspan')
plt.ylabel('Lognorm of FPKM avg of minspan')
plt.show()
```
