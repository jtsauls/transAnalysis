# Core metabolism reactions and their minspans

Taking iJO, what are the reactions of core metabolism and how do they distribute within the greater model and the minspans calcutlated for the greater model? Note, this is different that using a minspan calculation for the core model. The core reactions here are taken from: http://gcrg.ucsd.edu/Downloads/EcoliCore

Goals

* Find core metabolism reactions
* How connected are they? aka how many minspans are they in each?
* What are the minspans of core metabolism?
* What are their (both reactions and minspans as a whole) fpkm values?
* How does Eflux affect them?
* Output dicts for the visualization module.

Conventions

1. Use 'rxn' when iterating over reactions.
2. Use 'ms' when iterating over minspans.

```python
# Import a list of the core reaction with their subsystem
# Note I edited this file before import for reaction names
core_filename = filenameroot + 'ecolicore_rxns.csv'

core_rxns_df = pd.read_csv(core_filename, header=0)
HTML(core_rxns_df.to_html())
```

### How many minspans are these reactions involved in?

```python
# This code is from reaction_connectedness:
# Dict 1, minspan id (aka number, as key, list of reactons as values
minspan_ids = range(1, len(minspan_list)+1)
print('Number of minspans and Ids', 
      len(minspan_list), len(minspan_ids))

minspan_k_dict = {}
for k in minspan_ids:
    # The keys will one day be strings, so I will start it that way.
    minspan_k_dict[str(k)] = minspan_list[k-1]
    
print('Lenght of minspan_k_dict', len(minspan_k_dict))

# Dict 2, reaction name as k, with minspans involved as values.
reaction_k_dict = {}
reaction_wo_minspans = []
for reaction in model.reactions:
    temp_reaction_list = []
    for k, v in minspan_k_dict.iteritems():
        if reaction.id in v:
            temp_reaction_list.append(k)
    if len(temp_reaction_list) == 0:
        reaction_wo_minspans.append(reaction.id)
    reaction_k_dict[reaction.id] = temp_reaction_list
    
print('Length of reaction_k_dict, number of reactions in model', 
      len(reaction_k_dict), len(model.reactions))
print('Number of reactions w/o minspans', len(reaction_wo_minspans))
```

```python
# From core find the number of minspans per reaction
core_rxns_list = list(core_rxns_df.loc[:,'Reaction'])
rxn_proms = []
for rxn in core_rxns_list:
    try:
        rxn_proms.append(len(reaction_k_dict[rxn]))
    except:
        rxn_proms.append(float('nan'))
        print(rxn,'was not found')
    
rxn_proms = pd.Series(rxn_proms)

core_rxns_df['Promiscuity'] = rxn_proms
```

```python
# Print results
core_rxns_df.sort(columns='Promiscuity', ascending=False, 
                  inplace=True)
HTML(core_rxns_df.to_html())
```

### What are the minspans of core metabolism?

The effort here is to find which minspans are composed of all or nearly all core metabolism reactions. 

Relies on core_rxns_list, reaction_k_dict

```python
core_minspans_dict = {}

for rxn in core_rxns_list:
    for ms in reaction_k_dict[rxn]:
        if ms not in core_minspans_dict:
            core_minspans_dict[ms] = 1
        else:
            core_minspans_dict[ms] = \
                core_minspans_dict[ms] + 1
                
print('How many minspans these reactions are involved in',
      len(core_minspans_dict))
print('Total number of minspans', len(minspan_list))
```

### Core reactions are involed in many minspans

But are they the majority of many?

```python
k = core_minspans_dict.keys()
v = core_minspans_dict.values()
core_minspans_tup = zip(k, v)
core_minspans_tup.sort(key=lambda x: x[1], reverse=True)
minspans = []
core_rxn_count = []
total_rxn_count = []
fraction_core = []
rxn_list = []
for minspan in core_minspans_tup:
    minspans.append(minspan[0])
    core_rxn_count.append(minspan[1])
    total_rxn_count.append(len(minspan_k_dict[minspan[0]]))
    fraction_core.append(float(minspan[1]) / \
                         len(minspan_k_dict[minspan[0]]))
    rxn_list.append(minspan_k_dict[minspan[0]])
    
minspans = pd.Series(minspans, index=minspans)
core_rxn_count = pd.Series(core_rxn_count, index=minspans)
total_rxn_count = pd.Series(total_rxn_count, index=minspans)
fraction_core = pd.Series(fraction_core, index=minspans)
rxn_list = pd.Series(rxn_list, index=minspans)

core_minspan_df = {'Minspan' : minspans,
                   '# core rxns' : core_rxn_count,
                   '# total rxns' : total_rxn_count,
                   '% core rxns' : fraction_core,
                   'Reactions' : rxn_list}
columns = ['Minspan', '# core rxns', '# total rxns', '% core rxns',
           'Reactions']

core_minspan_df = pd.DataFrame(core_minspan_df,
                               columns=columns,
                               index=minspans)
core_minspan_df.sort(columns='% core rxns', ascending=False,
                      inplace=True)
HTML(core_minspan_df.to_html(float_format=lambda x: '%10.2f' % x))
```

### FPKM values for the reactions and minspans of core metabolism


```python
# Make the reaction to fpkm dict
fpkm_filename = filenameroot + 'bop27isoforms.fpkm_tracking' 
gene_fpkm_dict = ta.make_gene_fpkm_dict(fpkm_filename)
reaction_fpkm_dict = ta.make_reaction_fpkm_dict(model, 
                                                gene_fpkm_dict, 0)
```

```python
# Pull out fpkm values for core reactions
core_rxns_list = list(core_rxns_df.loc[:,'Reaction'])
rxn_fpkms = []
for rxn in core_rxns_list:
    try:
        rxn_fpkms.append(reaction_fpkm_dict[rxn])
    except:
        rxn_fpkms.append(float('nan'))
        print(rxn,'was not found')
    
rxn_fpkms = pd.Series(rxn_fpkms)
core_rxns_df['FPKM'] = rxn_fpkms
```
