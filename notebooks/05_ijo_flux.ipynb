## Minspan flux vs minspan transcript level

### Find flux per minspan
Here we are summing the abs(flux) for each reaction in a minspan, and then ranking the minspans based on their cumulative fluxes.

Alternatively, you can project the fluxes onto the minspans.

```python
# Dictionary with minspans id as key, reactions as values. 
minspan_k_dict = ta.make_minspan_k_dict(minspan_list)
```

```python
# This averages the fluxes per minspan. 
minspan_flux_list = []
for minspan in minspan_list:
    flux_temp = 0
    for rxn in minspan:
        flux_temp += abs(sol_pfba[rxn])
    flux_temp = flux_temp / len(minspan)
    minspan_flux_list.append((minspan, flux_temp))
```

```python
# Alternative, use the flux matrix
# Import the minspan flux matrix -- sligtlly different file, names adjusted
minspan_filename = filenameroot + 'MinSpanPathways_1.xlsx'
sheetname = 'iJO1366'
minspan_matrix = pd.read_excel(minspan_filename, sheetname,
                               index_col='Rxn', na_values=0)
minspan_matrix = minspan_matrix.drop('Pathway Length')
minspan_matrix = minspan_matrix.drop('Rxn Usage', 1)
minspan_matrix = minspan_matrix.fillna(0)

# Find consensus reactions... minspan.xls does not have them all
flux_rxns = xpfba_df.index.tolist()
minspan_rxns = minspan_matrix.index.tolist()
rxns_no_minspan = list(set(flux_rxns) - set(minspan_rxns))
xpfba_cut = xpfba_df.drop(rxns_no_minspan)
```

```python
xpfba_df.loc[rxns_no_minspan].sort(column='pFBA Flux', ascending=False).head()
```

```python
# Matrix multiplication and weird conversion
minspan_flux_list = np.dot(minspan_matrix.T, xpfba_cut)
minspan_flux_list = list(np.abs(minspan_flux_list.reshape(-1,)))
#minspan_flux_list = list(minspan_flux_list.reshape(-1,))
minspan_flux_list = zip(minspan_list, minspan_flux_list)
```

```python
print minspan_flux_list
```

```python
minspan_ids = range(1, len(minspan_list) + 1)
minspan_series = pd.Series(zip(*minspan_flux_list)[0], index = minspan_ids)
fluxes = np.array(zip(*minspan_flux_list)[1])
fluxes = [0 if flux > 10000 else flux for flux in fluxes]
minspan_flux_series = pd.Series(fluxes, index = minspan_ids)
minspan_flux_norm = pd.Series(fluxes / max(fluxes), index=minspan_ids)
minspan_flux_df = {'Reactions' : minspan_series,
                   'Flux' : minspan_flux_series,
                   'Flux norm' : minspan_flux_norm}
minspan_flux_df = pd.DataFrame(minspan_flux_df, columns=['Reactions', 'Flux', 'Flux norm'],
                               index = minspan_ids)
minspan_flux_df.sort(columns='Flux', ascending=True, inplace=True)
HTML(minspan_flux_df.to_html())
```

### Find transcript level per minspan
This relies on the transAnalysis module to go from an FPKM tracking file or microarray data to individual transcript levels per reaction, and then from that an average transcript level per minspan.

```python
# No cutoff
fpkm_filename = filenameroot + 'bop27_1_isoforms.fpkm_tracking'
gene_fpkm_dict = ta.make_gene_fpkm_dict(fpkm_filename)
reaction_fpkm_dict = ta.make_reaction_fpkm_dict(model, gene_fpkm_dict)
minspan_fpkm_list = ta.make_minspan_fpkm_list(minspan_list, reaction_fpkm_dict)
```

```python
# Alternative, calculate minspan_fpkm_list excluding highly connected reactions
reaction_k_dict = ta.make_reaction_k_dict(model, minspan_k_dict)

cutoff = 100
reaction_fpkm_dict_cut = reaction_fpkm_dict

for k, v in reaction_k_dict.iteritems():
    if len(v) >= cutoff:
        reaction_fpkm_dict_cut[k] = float('nan')

minspan_fpkm_list = ta.make_minspan_fpkm_list(minspan_list, 
                                              reaction_fpkm_dict_cut)
```

```python
minspan_ids = range(1, len(minspan_list) + 1)

# bop df
avgs = np.array(zip(*minspan_fpkm_list)[0])
minspan_avgs = pd.Series(avgs, index=minspan_ids)
minspan_reactions = pd.Series(zip(*minspan_fpkm_list)[1],
                                  index=minspan_ids)
minspan_norm = pd.Series(avgs / max(avgs), index=minspan_ids)
minspan_lognorm = np.log2(1 + ((avgs - min(avgs)) / (max(avgs) - min(avgs))))
minspan_lognorm = pd.Series(minspan_lognorm, index=minspan_ids)
fpkm_df = {'Reactions' : minspan_reactions,
          'FPKM avg' : minspan_avgs,
          'FPKM norm' : minspan_norm,
          'FPKM lognorm' : minspan_lognorm}
columns = ['Reactions', 'FPKM avg', 'FPKM norm', 'FPKM lognorm']
fpkm_df = pd.DataFrame(fpkm_df, columns=columns, index=minspan_ids)
```

### Compare

```python
# Addending the FPKM averages onto the flux df from above. 
minspan_flux_df['FPKM avg'] = minspan_avgs
```

```python
minspan_flux_df.sort(columns='Flux', ascending=False, inplace=True)
HTML(minspan_flux_df.to_html())
```

```python
reaction_levels = [0, 0]
reaction_levels[0] = fluxes / max(fluxes)
reaction_levels[1] = avgs / max(avgs)

# Correlation
pear_cor = stats.pearsonr(reaction_levels[0],
                          reaction_levels[1])
print("Pearson correlation, p-value", pear_cor)

spear_cor = stats.spearmanr(reaction_levels[0],
                            reaction_levels[1])
print("Spearman correlation, p-value", spear_cor)

# Plotting
plt.figure(figsize=(10,10), dpi=100)
# Value comparison.
plt.title('Minspan flux vs minspan transcript level')
plt.xlabel('Minspan flux')
plt.ylabel('Minspan transcript level') 
plt.scatter(reaction_levels[0], reaction_levels[1], c='g')
```
