#Comparing fpkms for reactions

Is there a lot of variation for fpkm values by reactions under different conditions?

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
from pickle import load # Can't get SBML to work, using pickle

# Have to direct to folder to find module
%cd /Users/jt/Desktop/palsson_rotation/transAnalysis/
import transAnalysis

filenameroot = \
    '/Users/jt/Desktop/palsson_rotation/transAnalysis/data/'
```

```python
# Get model.
model_filename = filenameroot + 'iJO1366.pickle'
f = open(model_filename, "rb")
model = load(f)
f.close()
```

```python
fpkm_filename = filenameroot + 'bop27_1_isoforms.fpkm_tracking' 
#fpkm_filename = filenameroot + 'anaerobe1_isoforms.fpkm_tracking'
#fpkm_filename = filenameroot + 'anaerobe2_isoforms.fpkm_tracking'
#fpkm_filename = filenameroot + 'rpoBisoforms.fpkm_tracking'

gene_fpkm_dict = transAnalysis.make_gene_fpkm_dict(fpkm_filename)
```

```python
def fpkm_rxn_comparison(fpkm_filename1, fpkm_filename2, model, mode=0):
    '''Graphs a comparison betweet to fpkm data for reactios'''
    
    filename_list = [fpkm_filename1, fpkm_filename2]
    reaction_levels = [0, 0]
    x = range(1,len(model.reactions)+1)
    
    for i in range(2):
        gene_fpkm_dict = transAnalysis.make_gene_fpkm_dict(
                         filename_list[i])
        reaction_fpkm_dict = transAnalysis.make_reaction_fpkm_dict(
                             model, gene_fpkm_dict, mode)
        
        reaction_fpkm_list = reaction_fpkm_dict.items()
        reaction_levels[i] = zip(*reaction_fpkm_list)[1]
        reaction_levels[i] = np.array([x if ~np.isnan(x) else 0 for 
                              x in reaction_levels[i]])
        a = min(reaction_levels[i])
        b = max(reaction_levels[i])
        reaction_levels[i] = np.log2(1 + 1023 * ((reaction_levels[i] - a) / (b - a)))
        
    # Correlation
    pear_cor = stats.pearsonr(reaction_levels[0],
                              reaction_levels[1])
    print("Pearson correlation, p-value", pear_cor)
    
    spear_cor = stats.spearmanr(reaction_levels[0],
                                reaction_levels[1])
    print("Spearman correlation, p-value", spear_cor)
    
    # Plotting
    plt.figure(figsize=(10,10), dpi=100)
    # Value comparison.
    plt.title('RNAseq vs. microarray levels for matched reactions')
    plt.xlabel('FPKM (log2 normalized)')
    plt.ylabel('mircroarray levels (log2 normalized)') 
    plt.scatter(reaction_levels[0], reaction_levels[1], c='g')
```

```python
fpkm_filename1 = filenameroot + 'an_isoforms.fpkm_tracking'
fpkm_filename2 = filenameroot + 'anaerobe2_isoforms.fpkm_tracking' 
fpkm_rxn_comparison(fpkm_filename1, fpkm_filename2, model, 1)
```

```python
fpkm_filename1 = filenameroot + 'ar_isoforms.fpkm_tracking'
fpkm_filename2 = filenameroot + 'bop27_2_isoforms.fpkm_tracking'
fpkm_rxn_comparison(fpkm_filename1, fpkm_filename2, model, 1)
```

```python
fpkm_filename1 = filenameroot + 'rpoBisoforms.fpkm_tracking'
fpkm_filename2 = filenameroot + 'bop27_2_isoforms.fpkm_tracking'
fpkm_rxn_comparison(fpkm_filename1, fpkm_filename2, model, 1)
```
